```python
ioc = load_io(io_file)
```

```python
def inp_out_val_to_str(ipv):
    inp = ipv[0]
    outp = ipv[1]
    steps = ipv[2]
    
    if steps == []:
        steps_str = "[]"
        print("zero")
    else:
        steps_str = "[" + ",".join(["'" + step + "'" for step in steps]) + "]"
        
    full_string = ", ".join(["'" + inp + "'", "'" + outp + "'", steps_str])
    
    return "[" + full_string + "]"
    
    
```

```python
fo = open("ioc.js", "w")
fo.write("var all_input_outputs = {\n")

for key in ioc:
    key_string = "(" + ", ".join([str(x) for x in key]) + ")"
    value = ioc[key]
    
    elt_strs = []
    
    for elt in value:
        elt_strs.append(inp_out_val_to_str(elt))
        
    fo.write("'" + key_string + "'" + ": " + "[" + ", ".join(elt_strs) + "],\n")
    
fo.write("}")
```
